Load the dataset into a Pandas DataFrame and display the first 5 rows.
Check the shape, column names, and summary statistics of the dataset.

```python
import pandas as pd

data = pd.read_csv("credit_card_transactions.csv")
data
```

```python
# Load the dataset into a Pandas DataFrame and display the first 5 rows

data.head()
```

```python
# Check the shape, column names, and summary statistics of the dataset

print('Shape of Data \n ',data.shape) # Row & Column
print('Column Name \n', data.columns.to_list()) #Featching Column Name
print('Summary Statistics \n ',data.describe())

# print('Summary Statistics \n ',data.describe(include='all')) #include='all' when we want summary of all column
```

```python
data.dtypes #Checking data type
```

```python
# Identify and handle missing values (fill or drop based on the data type)

print(data.isnull().sum()) #Checking for Null Values
```

```python
# Filling missing values in Amount Column

data['Merchant'].fillna('Unknown',inplace=True)
data['Category'].fillna('Unknown',inplace=True)
data['Payment_Mode'].fillna('Unknown',inplace=True)
data['Transaction_Status'].fillna('Unknown',inplace=True)
data['Transaction_Type'].fillna('Unknown',inplace=True)
```

```python
# Filling missing values in Amount Column

data['Merchant'] = data['Merchant'].fillna('Unknown')
data['Category'] = data['Category'].fillna('Unknown')
data['Payment_Mode'] = data['Payment_Mode'].fillna('Unknown')
data['Transaction_Status'] = data['Transaction_Status'].fillna('Unknown')
data['Transaction_Type'] = data['Transaction_Type'].fillna('Unknown')

data['Amount'] = data['Amount'].fillna(data['Amount'].median()) #mean()

print(data.isnull().sum())
print('Data Type \n', data.dtypes)
```

```python
# Convert Transaction_Date into datetime format and extract year, month, and day as new columns

data['Transaction_Date'] = pd.to_datetime(data['Transaction_Date']) #errors='coerce'

data.dtypes
```

```python
# extract year, month, and day as new columns

data['Year'] = data['Transaction_Date'].dt.year
data['Month'] = data['Transaction_Date'].dt.month 
data['Date'] = data['Transaction_Date'].dt.day

print(data.info())
```
